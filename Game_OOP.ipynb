```python
#Creating a class player with basic attributes
class Player:
    #The constructor must start with argument 'self', and later include every attribute
    def __init__(self, name, age, attack, defense):
        self.name = name
        self.age = age
        self.attack = attack
        self.defense = defense
    #Basic methods include all get and set methods for every attribute. I forgot to include in this class
    #The info method allows to display the attributes of an object
    def info(self):
        print(f'Player name: {self.name}\nAge: {self.age}\nAttack skills: {self.attack}\nDefense skills: {self.defense}')
```

```python
#Creating two objects. Please note the syntax of the object. variable = Class(particular attributes)
messi = Player('Lionel Messi', 34, 95, 78)
garcia = Player('Eric Garcia', 23, 67, 76)
```

```python
import random
#Created a class called Game where I bring to life interactions between object players
class Game:
    def __init__(self, player1, player2, score1, score2):
        self.player1 = player1
        self.player2 = player2
        self.score1 = score1
        self.score2 = score2
    #This method allows to create a game
    def scoring(self):
        time = 0
        play1 = self.player1
        play2 = self.player2
        goal_prob1 = play1.attack * (100 - play2.defense)/100**2
        goal_prob2 = play2.attack * (100 - play1.defense)/100**2
        
        while time <= 90:
            # Generate a random time interval for this chance
            time_interval = random.randint(1, 6)
            time += time_interval
            
            # Player 1 is attacking
            print(f'{play1.name} is attacking right now...{time} minute')
            if random.random() < goal_prob1:
                self.score1 += 1
                print(f'Goaaaaal')
                print(f'{play1.name} {self.score1} - {play2.name} {self.score2}')
            else:
                print('Goalkick')
            
            # Generate a random time interval for this chance
            time_interval = random.randint(1, 6)
            time += time_interval
            
            # Player 2 is attacking
            print(f'{play2.name} is attacking right now...{time} minute')
            if random.random() < goal_prob2:
                self.score2 += 1
                print(f'Goaaaaal')
                print(f'{play1.name} {self.score1} - {play2.name} {self.score2}')
            else:
                print('Goalkick')
                
        print(f'Final Blow!\nFinal Result: {play1.name} {self.score1} - {play2.name} {self.score2}')
```

```python
#The class will not create a real game unless I create an object game. In this examples I made a game 1 obje
game1 = Game(messi, garcia,0,0)
game1.scoring()
```
