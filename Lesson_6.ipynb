# Lesson 6: Repetition Structures - for loop

In Python, a repetition structure, also known as a loop, is used to execute a block of code multiple times. There are two types of loops in Python: the for loop and the while loop.

### Count-controlled loop




```python
for variable in [value 1, value 2, value 3, etc]:
    action 1
    action 2
    etc.
    
```

```python
print('I will display the numbers 1 through 5.') #title
for num in [1, 2, 3, 4, 5, 10]: 
    print(num)
```

```python
print('I will display some of the best football teams in the world.') #title
for teams in ['Arsenal','Barcelona', 'AC Milan', 'River Plate']: 
    print(teams)
```

Using the `range` function with the `for` loop

`range(start (include), end (exclude), step)`

```python
for x in range(0,24,5):
    print(x)
```

### Example 1. 
Write a loop that prints a table of the first ten numbers and their equivalent squares displayed in two columns.
  


```python
# Example 1
print(f'Number\tSquare')
print(f'**************')

for number in range(1,11,1):
    square = number **2
    print(f'{number}\t{square}')
```

### Try it yourself. 

Most of the world uses the Metric System for physical measurements. When international students come to study at USA, they struggle with converting from kilometers/hr to miles/hr. Create a chart that compares KPH with MPH. Display speeds from 40 KPH to 150 KPH using increments of 10. Formula: MPH = KPH / 1.609. (Use 1 floating number on display)

```python
print(f'KPH\tMPH')
print(f'**************')

for kph in range(40,151,10):
    mph = kph / 1.609
    print(f'{kph}\t{mph:.1f}')
```

## Nested Loops

A nested loop is a series of loops that are inside other loops. 

```python
for minutes in range(60):
    for seconds in range(60):
        print(f'{minutes:02d} : {seconds:02d}')
```

Exercise Monday:

The OCU Dinning Hall has commissioned you to develop a small app where students can order specific products from their phones. The idea is to display a standard menu option where student choose what they want to order and displays a total.

The menu should display the following:

“Welcome to the Mighty Oaks Cafeteria. Use the list numbers from the menu to make your order:

1.       Mighty Oaks Burger: $12

2.       Cheese Burger: $9

3.       French fries: $5

4.       Soda: $3

After ordering an item, the user should be asked if they want to order something else: Yes =1, No=0.

If the user wishes to order something else, the menu should display again with the prompt. If the user decides to finish the order. The total should be displayed.



```python
total = 0
prompt = 1
print(f'“Welcome to the Mighty Oaks Cafeteria.\n Use the list numbers from the menu to make your order:')

while prompt == 1:
    print(f'Use the list numbers from the menu to make your order:\n1. Mighty Oaks Burger: $12\n2. Cheese Burger: $9\n3. French fries: $5\n4. Soda: $3:')
    order = int(input("Enter your option (1-4): "))
    if order == 1:
        total = total + 12
    elif order == 2:
        total = total + 9
    elif order == 3:
        total = total + 5
    elif order == 4:
        total = total + 3
    else:
        print('You input an invalid option')
    
    prompt = int(input(f'Do you want to order something else (Yes = 1, No = 0)? '))
print(f'Your total is: ${total:.2f}\nThanks for your preference :)')
```
