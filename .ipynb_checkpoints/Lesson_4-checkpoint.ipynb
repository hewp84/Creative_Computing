## Comparison operators
!


```python
print(3 > 2)     # True
print(3 >= 2)    # True
print(3 < 2)     # False
print(2 < 3)     # True
print(2 <= 3)    # True
print(3 == 2)    # False
print(3 != 2)    # True
print('mango' == 'avocado')  #False
print('mango' != 'avocado')  # True
print('mango' < 'avocado')   # False
print('Mark' != 'Mary')      # True
print('Mark' == 'Mary')      # False
print('tomato' == 'potato')  # False
print('python' > 'dragon')   # True
```

Python offers reserved words as comparison operators:

* in: Returns True if the queried list contains a certain item(x in y)
* not in: Returns True if the queried list doesn't have a certain item(x in y)

```python
          # 
print('H in Hector', 'H' in 'Hector') # True
print('X in Hector', 'X' in 'Hector') # False
print('end' in 'Friend') # True
print('a in an:', 'a' in 'an')      # True
```

## Logical Operators
Logical operators are useful as combinations of comparison operators. Using their truth tables, their output is boolean.
!

```python
print(3 > 2 and 4 > 3) # True AND True = True
print(3 > 2 and 4 < 3) # True AND False = False
print(3 < 2 and 4 < 3) # False AND False = False
print(3 > 2 or 4 > 3)  # True OR True = True
print(3 > 2 or 4 < 3)  # True OR False = True

print(not 3 > 2)     # NOT TRue = False
print(not True)      # False
print(not False)     # True
print(not not True)  # True
print(not not False) # False
```

# Decision Structures and Conditionals

By default, statements in Python script are executed sequentially from top to bottom. If the processing logic require so, the sequential flow of execution can be altered in two way:

* Conditional execution: a block of one or more statements will be executed based on a true or false decision.
* Repetitive execution: a block of one or more statements will be repetitively executed as long as a certain expression is true.

## If-else statement



```python
if condition:
    action 1
    action 2
    action 3
    etc.
else:
    action 4
    action 5
    action 6
```

```python
#Write a script for the temperature flow chart
temperature = 20
if temperature  < 40:
    print(f'A little cold, isn\'t it')
else:
    print(f'Nice weather we\'re having.')
```
